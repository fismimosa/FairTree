```python
os.chdir('~\Datasets')
print(os.getcwd())
def load_bank(data_dir=''):

    
    
    
    _path = 'bank-additional-full.csv'
    data_path = os.path.join(data_dir, _path)
      
    df = pd.read_csv(data_path, sep=';', header=[0])
    print(df.head())
    print(df.columns)
    col = ['age','duration','euribor3m', 'nr.employed', 'cons.price.idx', 'campaign','marital']
    print(df.columns)
    col_index =[]
    i = 0
    for c in df.columns:
        if c in col:
            col_index.append(i)
        i=i+1
        
    df = df.iloc[:,col_index]#[0,2,5,11]]
    df1 = df.loc[df['marital'] == 'married']
    df2 = df.loc[df['marital'] == 'single']
    df3 = df.loc[df['marital'] =='divorced']
    
    
    df = [df1, df2, df3]
    df = pd.concat(df)
    n = df.shape[0]
    
    sex = df['marital']
    #sens_attributes = list(set(sex.astype(str).values))  
    sex_num = np.zeros(n, dtype=int)
    sex_num[sex.astype(str).values == 'married'] = 1  
    sex_num[sex.astype(str).values == 'single'] = 2
    sex_num[sex.astype(str).values == 'divorced'] = 3
   
    print(df)
    df['type'] = sex_num 
    
    print(df)
    
    df= df.drop(['marital'], axis=1)
    print(df)
    
    type_col = df['type']
    df= df.drop(['type'], axis=1)
    x = df.values  # returns a numpy array
    x_scaled = preprocessing.scale(x,axis=0)
      #  x_scaled = min_max_scaler.fit_transform(x)
    df = pd.DataFrame(x_scaled)
    print(df.head())
    x = df.values
    feanorm = np.maximum(1e-14, np.sum(x**2, axis=1))
    x_norm = x/ (feanorm[:,None]**0.5)
    df = pd.DataFrame(x_norm)

    df['type'] = type_col.values
    
    print(df)
    df.to_csv('bank-full_p_6col.csv',index=False)
        
        
    
load_bank()

```
